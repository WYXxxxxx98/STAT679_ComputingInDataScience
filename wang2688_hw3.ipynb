```python
#1.1
def duplicate_lines(infile, outfile, n):
    #checking that either of infile or outfile are not strings
    if not isinstance(infile,str) or not isinstance(outfile,str):
        raise TypeError('The infile and outfile should be strings.')
    # checking n is not a non-negative integer
    if not isinstance(n,int):
        raise TypeError('The n should be integer.')
    if n<0:
        raise ValueError('The n should be non-negative.')
        
    with open(infile,'r') as Fin:
        with open(outfile,'w') as Fout:
            if n==0:
                Fout.write('')
            for line in Fin:
                if '\n' in line:
                    pass
                else:
                    line=line+'\n'
                Fout.write(line*n)
```

```python
#1.2
def duplines_storage(infile, outfile, n):
    with open(infile,'r') as Fin:
        lines=list()
        for line in Fin:
            if '\n' in line:
                pass
            else:
                line=line+'\n'
            lines.append(line)
            
    with open(outfile,'w') as Fout:
        if n==0:
            Fout.write('')
        for line in lines:
            Fout.write(line*n)
```

```python
def duplines_direct(infile, outfile, n):
    with open(infile,'r') as Fin:
        with open(outfile,'w') as Fout:
            if n==0:
                Fout.write('')
            for line in Fin:
                if '\n' in line:
                    pass
                else:
                    line=line+'\n'
                Fout.write(line*n)
```

For option1, if the infile is large, the lines(line list) will be large. This might waste time for storage lines as the list.

```python
#1.3
def generate_test_file(filename,nlines,nchars):
    #import packages
    import random
    import string
    #checking filename is a string
    if not isinstance(filename,str):
        raise TypeError('The filename should be a string.')
    #checking nlines and nchars are int
    if not isinstance(nlines,int) or not isinstance(nchars,int):
        raise TypeError('The nlines and nchars should be integers.')
    #checking non-negative
    if nlines<0 or nchars<0:
        raise ValueError('The nlines and nchars should be non-negative.')
    
    with open(filename,'w') as file:
        if nlines==0:
            file.write('')
        if nchars==0:
            for i in range(nlines-1):
                file.write('\n')
        for i in range(nlines):
            for j in range(nchars):
                char=random.randint(0,25)
                new=string.ascii_lowercase[char]
                file.write(new)
            file.write('\n')
```

```python
#1.4
def time_trial(nlines,nchars,ntimes):
    #import packages
    import time
    
    #generating files
    generate_test_file('test_file.txt',nlines,nchars)
    
    #time for duplines_storage
    start_time1=time.time()
    duplines_storage('test_file.txt','test_out1.txt',ntimes)
    end_time1=time.time()
    time1=end_time1-start_time1
    
    #time for duplines_direct
    start_time2=time.time()
    duplines_direct('test_file.txt','test_out2.txt',ntimes)
    end_time2=time.time()
    time2=end_time2-start_time2
    
    return (time1,time2)
```

```python
#Again, your function should ignore punctuation, spaces, newlines and capitalization.
def count_bigrams_in_file(filename):
    #import package
    import string
    
    #create a new dictionary
    dic=dict()
    word_list=tuple()
    #open the file
    try:
        with open(filename,'r') as file:
            for line in file.readlines():
                line=line.lower()
                line=line.strip('\n')
                for word in line.split():
                    word_list=word_list+(word.strip(string.punctuation),)
                    
            for i in range(len(word_list)-1):
                if dic.get(word_list[i:i+2]):
                    dic[(word_list[i:i+2])]+=1
                else:
                    dic[(word_list[i:i+2])]=1
                        
    except FileNotFoundError:
        print('The file cannot be found.')
    except OSError:
        print('The argument should be a string.')
    return dic
```

```python
#2.3
def collocations1(filename):
    import string
    #set new variables
    wordlist=tuple()
    dic=dict()
    
    #spliting words
    try:
        with open(filename,'r') as file:
            for line in file:
                line=line.lower()
                for word in line.split():
                    wordlist=wordlist+(word.strip(string.punctuation),)

                    #checking whether the new word is in the set
                    n=len(wordlist)
                    if n==1:
                        dic[wordlist[0]]=set()
                    if n==2:
                        dic[wordlist[0]]={wordlist[1]}
                        dic[wordlist[1]]={wordlist[0]}
                    if n>2:
                        if dic.get(wordlist[n-2]):
                            if wordlist[n-3] not in dic[wordlist[n-2]]:
                                dic[wordlist[n-2]].add(wordlist[n-3])
                            if wordlist[n-1] not in dic[wordlist[n-2]]:
                                dic[wordlist[n-2]].add(wordlist[n-1])
                        else:
                            dic[wordlist[n-2]]={wordlist[n-3],wordlist[n-1]}

                l=len(wordlist)
                if dic.get(wordlist[l-1]):
                    if wordlist[l-2] not in dic[wordlist[l-1]]:
                        dic[wordlist[l-1]].add(wordlist[l-2])
                else:
                    dic[wordlist[l-1]]={wordlist[l-2]}
    except FileNotFoundError:
        print('The file cannot be found.')
    except OSError:
        print('The argument should be a string.')
    return dic
```

```python
import string
file_name='hw3example.txt'
def count_chars(file_name):
    character_counts=0
    with open(file_name,'r')as f:
        for line in f:
            character_counts+=len(line)
    return character_counts
def count_words(file_name):
    word_counts =0
    with open(file_name,'r')as f:
        for line in f:
            words = line.split()
            word_counts+=len(words)
    return word_counts
def count_lines(file_name):
    line_counts =0
    with open(file_name,'r')as f:
        for line in f:
            line_counts+=1
    return line_counts
```
