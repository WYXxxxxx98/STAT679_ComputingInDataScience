```python
#1.1
def get_page_title(s):
    if not isinstance(s,str):
        raise TypeError('The argument should be a string.')
    try:
        html=urllib.request.urlopen(s)
    except HTTPError:
        print('The argument cannot be read.')
        return
    except:
        raise ValueError('The argument should be a URL.')
        return
    soup = BeautifulSoup(html, 'html.parser')
    title=soup.title
    if title==None:
        return None
    return title.get_text()
```

```python
#1.2
def count_links(s):
    if not isinstance(s,str):
        raise TypeError('The argument should be a string.')
    try:
        html=urllib.request.urlopen(s)
    except HTTPError:
        print('The argument cannot be read.')
        return
    except:
        raise ValueError('The argument should be a URL.')
        return
    soup=BeautifulSoup(html,'html.parser')
    tag=[tag for tag in soup.find_all('a')]
    new_tag=[]
    for item in tag:
        if 'href' in item.attrs:
            new_tag.append(item)
    return len(new_tag)  #file or pic???
```

There is one child for tag table called tbody. In tbody, there are 17 children with name tr which represent each row in the table. Within each tr, there are children called th which represent the columns in the table. In several th, there are children with tag a which represent hyper links in the table.

```python
#2.2
def retrieve_climate_table(url):
    if not isinstance(url,str):
        raise TypeError('The argument should be a string.')
    try:
        html=urllib.request.urlopen(url)
    except HTTPError:
        print('The argument cannot be read.')
        return
    except:
        raise ValueError('The argument should be a URL.')
        return
    
    html_soup=BeautifulSoup(html,'html.parser')
    
    climate=[]
    for table in html_soup.find_all(text=re.compile('Climate data for')):
        climate.append(table.findParent('table'))
    
    if climate==[]:
        return None
    else:
        return BeautifulSoup(str(climate[0]),'html.parser')
```

```python
retrieve_climate_table('https://en.wikipedia.org/wiki/Los_Angeles')
```

```python
#2.3
def is_wiki_page( url ):
    if '/wiki/' not in url:
        return False
    else:
        return True

def list_climate_table_row_names(url):
    #checking URL can be open
    if not isinstance(url,str):
        raise TypeError('The argument should be a string.')
    try:
        html=urllib.request.urlopen(url)
    except HTTPError:
        print('The argument cannot be read.')
        return
    except:
        raise ValueError('The argument should be a URL.')
        return
    
    html_soup=BeautifulSoup(html,'html.parser')
    #checking URL is wiki
    if is_wiki_page(url)==True:
        pass
    else:
        raise ValueError('The URL is not a wiki page.')
    
    climate=[]
    for table in html_soup.find_all(text=re.compile('Climate data')):
        climate.append(table.findParent('table'))
    
    if climate==[]:
        return None
    else:
        row_name=[]
        for name in BeautifulSoup(str(climate[0]),'html.parser').find_all('th',scope='row'): 
            #tr: all content in table, th: the column and row head, td: content in table exclude index
            row_name.append(name.get_text().replace('\n',''))
    return row_name
```

```python
list_climate_table_row_names('https://en.wikipedia.org/wiki/Shanghai')
```

```python
#2.4
def get_data_in_row(url,name):
    #Retrieve the climate table:
    table=retrieve_climate_table(url)
    #Do not need error checking, it is included in retrieve_climate_table function in 2.2
    
    #Check whether the input row name in table:
    row_names=list_climate_table_row_names(url)
    if name not in row_names:
        raise ValueError('The input row name is not in the table.')
    elif name==row_names[0]:
        print('The input name is name of column name.')
    else:
        #Find the row name in table:
        x=table.find_all(text=name+'\n')[0]
        row=BeautifulSoup(str(x.findParent('tr')),'html.parser') #find the all content with this row name
        #Re-store the data into a new list:
        data=[]
        for item in row.find_all('td'):
            data.append(item.get_text().replace('\n',''))
        return data
```

```python
get_data_in_row('https://en.wikipedia.org/wiki/Shanghai','Month')
```
