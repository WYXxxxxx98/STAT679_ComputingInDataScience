```python
#1.1-1.9 excpet 1.3/1.4
class Vector:
    def __init__(self,dim,entry=tuple()):
        self.set_dim(dim)
        self.set_entry(entry)
    
    #for initial values in init function
    def set_dim(self,dim):
        if not isinstance(dim,int):
            raise TypeError('The dimension should be an integer.')
        if dim<=0:
            raise ValueError('The dimension should be positive.')
        self.dim=dim
        
    def set_entry(self,entry):
        if not isinstance(entry,(list,tuple)):
            raise TypeError('The entry should be a list or tuple.')
        for item in entry:
            if not isinstance(item,(int,float)):
                raise TypeError('The items in entry should be float or integer.')
        if len(entry)==0:
            self.entry=(0,)*self.dim
        elif len(entry)!=self.dim:
            raise ValueError('The dimension and the number of supplied entries disagree.')
        else:
            self.entry=entry
            
    #get variables       
    def get_dim(self):
        return self.dim
    def get_entries(self):
        return self.entry
    
    #comparison
    def __eq__(self,other):
        if self.dim!=other.dim:
            raise ValueError('The vectors comparied should have same dimension.')
        return self.entry==other.entry
    
    def __ne__(self,other):
        if self.dim!=other.dim:
            raise ValueError('The vectors comparied should have same dimension.')
        return self.entry!=other.entry
    
    def __lt__(self,other):
        if self.dim!=other.dim:
            raise ValueError('The vectors comparied should have same dimension.')
        return self.entry<other.entry
    
    def __le__(self,other):
        if self.dim!=other.dim:
            raise ValueError('The vectors comparied should have same dimension.')
        return self.entry<=other.entry
    
    def __gt__(self,other):
        if self.dim!=other.dim:
            raise ValueError('The vectors comparied should have same dimension.')
        return self.entry>other.entry
    
    def __ge__(self,other):
        if self.dim!=other.dim:
            raise ValueError('The vectors comparied should have same dimension.')
        return self.entry>=other.entry
    
    ##adding and substracting
    def __add__(self, other):
        if self.dim != other.dim:
            raise ValueError('The vectors added should have same dimension.')
        new = tuple()
        for i in range(self.dim):
                new=new+(self.entry[i] + other.entry[i],)
        return (Vector(self.dim, new))
    
    def __sub__(self, other):
        if self.dim != other.dim:
            raise ValueError('The vectors substracted should have same dimension.')
        new = tuple()
        for i in range(self.dim):
                new=new+(self.entry[i] - other.entry[i],)
        return (Vector(self.dim, new))
    
    ## inner production
    def dot(self, other):
        if not isinstance(other,Vector):
            raise TypeError('The input should be vectors.')
        if self.dim != other.dim:
            raise ValueError('The vectors for inner production should have same dimension.')
        answer=0
        for i in range(self.dim):
                answer+=self.entry[i] * other.entry[i]
        return answer
    
    # scalar multiplication
    def __mul__(self, other):
        new=tuple()
        if isinstance(other,(int,float)):
            for i in range(self.dim):
                new=new+(self.entry[i]*other,)
        elif isinstance(other,Vector):
            if self.dim != other.dim:
                raise ValueError('The vectors substracted shoud have same dimension.')
            for i in range(self.dim):
                    new=new+(self.entry[i]*other.entry[i],)
        else:
            raise TypeError('The input should be a Vector, a float or an integer.')
        return (Vector(self.dim, new))
    
    def __rmul__(self, other):
        new=tuple()
        if isinstance(other,(int,float)):
            for i in range(self.dim):
                new=new+(self.entry[i]*other,)
        return (Vector(self.dim, new))
                
    #p-norm function
    def norm(self,p):
        norm=0
        if not isinstance(p,(int,float)):
            raise TypeError('The input should be an integer or a float.')
            
        if p<0:
            raise ValueError('The input should be non-negative.')
        elif p==0:
            for i in range(self.dim):
                if self.entry[i]!=0:
                    norm+=1
        elif p==float('inf'):
            for i in range(self.dim):
                a=[abs(item) for item in self.entry]
                norm=max(a)
        else:
            for i in range(self.dim):
                norm+=abs(self.entry[i])**p
            norm=norm**(1/p)
        return norm
```

```python
#2.1
class SymIntRW:
    def __init__(self):
        self.current_position=0
        self.history=[]
        
    def get_time(self):
        return len(self.history)
    
    def get_history(self):
        return self.history
    
    def get_position(self):
        return self.current_position
    
    def step(self):
        self.history.append(self.current_position)
        import random
        direction=[-1,1]
        self.current_position+=direction[random.randint(0,1)]
        return self
```

```python
#2.2
class IntegerRW(SymIntRW):
    def __init__(self,p):
        self.current_position = 0
        self.history = []
        self.set_prob(p)

    def set_prob(self,p=0.5):
        if not isinstance(p,(int,float)):
            raise TypeError('The probability should be an interger or a float.')
        if p>1 or p<0:
            raise ValueError('The probability should be in range [0,1].')
        self.p=p
    
    def step(self):
        self.history.append(self.current_position)
        import random
        prob=random.random()
        if prob<=self.p:
            self.current_position+=1
        else:
            self.current_position-=1
        return self
```

```python
class SSRW(IntegerRW):
    def __init__(self,d):
        self.set_dimension(d)
        self.current_position = [0] * self.dimension
        self.history = []
        
    def set_dimension(self,d = 1):
        if not isinstance(d,int):
            raise TypeError('Dimension should be an integer.')
        if d<=0:
            raise ValueError('Dimension should be positive.')
        self.dimension=d
        
    def step(self):
        import random
        self.history.append(tuple(self.current_position))
        j=random.randint(0,self.dimension-1)
        direction=random.choice([-1,1])
        self.current_position[j]+=direction

        return self
```
